```python
Y_pred_test=logreg.predict(X_test)
```

```python
logreg.predict_proba(X_test)[:,1]
```

```python
# Check accuracy score
from sklearn.metrics import accuracy_score
print("Model Accuracy:",accuracy_score(Y_test,Y_pred_test))
```
